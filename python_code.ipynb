```python
import pandas as  pd
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
```

```python
# Distribution in original data (count any column values, but don't take columns which contain NULL values, otherwise it will give wrong count)

data.groupby('Potability').count()['Solids'].reset_index().sort_values(by='Potability',ascending=False)
```

```python
#Three columns contain NULL values: ph, Sulfate and Trihalomethanes
```

```python
# Drop columns where at least two of these three coumns are NULL.
# If only one column is NULL, replace it with average.

data = data.drop(data[(data.ph.isna()) & (data.Sulfate.isna()) & (data.Trihalomethanes.isna())].index)
data = data.drop(data[(data.Sulfate.isna()) & (data.Trihalomethanes.isna())].index)
data = data.drop(data[(data.ph.isna()) & (data.Sulfate.isna())].index)
data = data.drop(data[(data.ph.isna()) & (data.Trihalomethanes.isna())].index)
data.fillna(data.mean(), inplace=True)
```

```python
# Split data into training and testing

train_data, test_data = train_test_split(data, test_size=0.2,  shuffle=True)
train_data = train_data.reset_index()
test_data = test_data.reset_index()
```

### NOTE: 

For imbalanced dataset: normal ML methods like logistic regression, k-nearest neighbours and SVM can perform bad. <br>
But, decision tree and ensemble methods do not get affected by imbalance dataset much. <br>
So, for sure, the best working model for our dataset will be Random forest or gradient boosting or Adaboost.<br>

#### Try using k-fold cross validation.

```python
y_train = train_data['Potability']
x_train = train_data.drop(columns=['Potability'])

y_test = test_data['Potability']
x_test = test_data.drop(columns=['Potability'])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score,f1_score, confusion_matrix,classification_report
```

```python
# Logistic Regression

lr = LogisticRegression(C=1.0)
lr.fit(x_train, y_train)
pred_lr = lr.predict(x_test)
```

```python
print(confusion_matrix(y_test,pred_lr))
print(accuracy_score(y_test,pred_lr))
print(classification_report(y_test,pred_lr))
```

```python
# Results are not good with Logistic Regression, let's run different models for the same training testing data and compare them.
```

```python
accuracy = []
f1_s = []
true_neg = []
false_neg= []
false_pos = []
true_pos = []
```

```python
# Print accuracy of model
def predict(model,X_train,y_train,X_test,y_test):
    
    clf = model
    clf.fit(X_train, y_train)
    predictions=clf.predict(X_test)
    confusion_matrix(y_test,predictions)
    print("{}" .format(model))
    print(classification_report(y_test,predictions))
    print('-'*50)
    cm = confusion_matrix(y_test,predictions)
    print(cm)
    cm = list(cm)
    true_neg.append(cm[0][0])
    false_pos.append(cm[0][1])
    false_neg.append(cm[1][0])
    true_pos.append(cm[1][1])
    acc = accuracy_score(y_test,predictions)
    accuracy.append(acc)
    print('Accuracy', acc)
    f1 = f1_score(y_test,predictions, average='macro')
    f1_s.append(f1)
    print('F1',f1)
    print('-'*50)
    
   
```

```python
# Create a list of the regression models to be used
models=[LogisticRegression(C=1.0),SVC(),MultinomialNB(),KNeighborsClassifier(n_neighbors=5),
        DecisionTreeClassifier(), RandomForestClassifier()]
```

```python
# Loop through the list of models and use function to trian and make predictions
for model in models:
    predict(model,x_train, y_train, x_test ,y_test)
```

For imbalance dataset: <br>
There are two major approaches- 
1) Use ensemble methods (Random Forest/ Boosting) <br>
2) Use oversampler or undersampler

```python
#pip install imblearn
#pip install delayed
```

```python
from imblearn.over_sampling import RandomOverSampler
from collections import Counter
```

```python
os=RandomOverSampler(sampling_strategy=0.999)
x_train_os,y_train_os= os.fit_resample(x_train,y_train)
print("The number of classes before fit {}".format(Counter(y_train)))
print("The number of classes after fit {}".format(Counter(y_train_os)))
```

```python
# Loop through the list of models and use function to trian and make predictions
for model in models:
    predict(model,x_train_os, y_train_os, x_test ,y_test)
```

```python
# Create dataframe with all the results for different models (In model_names below, OS represent Oversampler method)

model_names = ['Logistic', 'SVC', 'Multinomial', 'k-nearest-classifier', 'Decision Tree', 'Random Forest', 'Logistic_OS', 'SVC_OS', 'Multinomial_OS', 'k-nearest-classifier_OS', 'Decision Tree_OS', 'Random Forest_OS']

df = pd.DataFrame(list(zip(accuracy, f1_s, true_neg, false_neg, false_pos, true_pos)),
               columns =['Accuracy','F1 score', 'True Negative', 'False Negative', 'False Positive', 'True Positive'])
```

```python
df
```

```python
df_t = df.T                               # Transpose of dataframe
df_t = df_t.set_axis(model_names, axis=1, inplace=False)   
df_compare = df_t[['Logistic', 'Logistic_OS', 'SVC', 'SVC_OS', 'Multinomial', 'Multinomial_OS', 'k-nearest-classifier', 'k-nearest-classifier_OS','Decision Tree', 'Decision Tree_OS', 'Random Forest','Random Forest_OS' ] ]
```
